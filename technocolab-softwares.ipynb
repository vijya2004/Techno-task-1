# Data preprocessing
replacing target variable with numeric value


```python
df_real.Attrition.replace({"Yes":1,"No":0}, inplace=True)
```

deleting the following features becaues it has same value in all observation

```python
df_real.drop(columns=["EmployeeCount", "StandardHours"], inplace=True)
```

```python
df_real
```

```python
#finding unique values

unique_values_1 = set(df_real.Department)             # this give only unique values
unique_values_2 = df_real.Department.value_counts()   # this give unique values with total count/occurance i.e, frequency

print(unique_values_1)
print(unique_values_2)
```

saperating numerical and catagorical values

```python
numerical_columns =df_real.select_dtypes(include=np.number).columns
catagorical_columns =df_real.select_dtypes(exclude=np.number).columns
```

```python
catagorical_columns
```

```python
for i in catagorical_columns:
    print(f"\n====================> {i} \n")
    print(df_real[i].value_counts())
```

```python
numerical_columns
```

```python
for i in numerical_columns:
    print(f"\n====================> {i} \n")
    print(df_real[i].value_counts())
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_real[numerical_columns] = scaler.fit_transform(df_real[numerical_columns])
df_real
```

creating the dummy variable for catagorical values

```python
catagorical_dummy = pd.get_dummies(df_real[catagorical_columns], drop_first=True)

catagorical_dummy = catagorical_dummy.astype(bool)
```

```python
catagorical_dummy.info()
```

```python
df_sec = pd.concat([df_real[numerical_columns], catagorical_dummy], axis=1)
```

```python
df_sec
```

MODEL CREATION

```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import classification_report,accuracy_score
```

dividing x and y features

```python
x = df_sec.drop(columns="Attrition")
y = df_sec.Attrition
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=10)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=1000)
```

```python
model.fit(x_train, y_train)
```

```python
x_pred = model.predict(x_train)
```

```python
y_pred = model.predict(x_test)
```

```python
metrics.confusion_matrix(y_train, x_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')

print(classification_report(y_test, y_pred))
```
